## Create a Python program that:

### 1- Asks the user to enter two numbers.

### 2-Stores them as variables (make sure to convert them to integers).

### 3-Performs and prints the results of the following operations:

* ## Addition

* ## Subtraction

* ## Multiplication

* ## Division

* ## Power (num1 ** num2)

* ## Modulus (num1 % num2)

* ## Floor division (num1 // num2)






```python
num1 = int(input("enter first number: "))
num2 = int(input("enter second number: "))

x = num1
y = num2

z = x + y
r = x - y
e = x * y
w = x / y
d = x ** y
p = x % y
l = x // y

print(z, r, e, w, d, p, l)
```

# Write a Python program that does the following:

# Asks the user to enter the length and width of a rectangle (Float).

# Calculates the area and perimeter of the rectangle(Float).

# Prints the result neatly, for example:
# Enter length: 5
# Enter width: 3
# Area: 15.0
# Perimeter: 16.0

```python
length = float(input("Enter length: "))
width = float(input("Enter width: "))
x=length
y=width
area=x*y
perimeter=2*(x*y)
print(area,perimeter)

```
